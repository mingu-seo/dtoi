```python
def getPortionValue(keyword):    
    word = urllib.parse.quote(keyword)
    url = 'https://www.fatsecret.kr/%EC%B9%BC%EB%A1%9C%EB%A6%AC-%EC%98%81%EC%96%91%EC%86%8C/%EC%9D%BC%EB%B0%98%EB%AA%85/'+word

    req = urllib.request
    res = req.urlopen(url)
    data = res.read()
    text = data.decode('utf-8')

    
    #text = requests.get(url, data=param).text
    text = requests.get(url).text

    soup = BeautifulSoup(text)
    
    portionID = 0

    for e in soup.select('#updateForm > table > tr.norow > td:nth-child(2) > table > tr > td:nth-child(2) > select > option'):
        if e.get_text().strip() == 'g' : portionID = e.attrs['value']
    
    return portionID
```
